```python
import pandas as pd
import unicodedata
```

```python

def clean_column_name(col):
    col = ''.join(c for c in unicodedata.normalize('NFKD', col) if not unicodedata.combining(c))
    col = col.lower()
    col = col.replace("'", "").replace("-", "_").replace(" ", "_")
    col = ''.join(char if char.isalnum() or char == '_' else '' for char in col)
    col = '_'.join(filter(None, col.split('_')))
    return col
```

```python
def clean_csv(input_file, output_file):
    df = pd.read_csv(input_file, encoding="latin1")  
    df.columns = [clean_column_name(col) for col in df.columns]
    df.to_csv(output_file, index=False, encoding="utf-8")
    print(f"Cleaned file saved as: {output_file}")
    print("New Columns:", df.columns.tolist())

clean_csv("Hypermarche.csv", "Hypermarche_clean.csv")
clean_csv("Objectifs_ventes.csv", "Objectifs_ventes_clean.csv")
```
